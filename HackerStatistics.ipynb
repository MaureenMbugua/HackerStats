```python
#import numpy package
import numpy as np

#initialize our random walk
random_walk=[0]


#control to determine next move in our random walk
#simulate 100 steps in our random walk

for x in range(100): 
 steps = random_walk[-1]
    #simulate a dice roll
 dice_roll = np.random.randint(1,7)
 if dice_roll<=2:
    steps=max(0, steps-1) #to ensure you don't go below zero
 elif dice_roll<=5:
    steps=steps+1
 else:
    steps=steps+(np.random.randint(1,7))
    #add the next step to our random walk
 random_walk.append(steps)
    
 
    
print(random_walk)
```

```python
#visualise our random walk
import matplotlib.pyplot as plt

plt.plot(random_walk)
plt.title('One Random Walk with 100 steps')
plt.show()
```

```python
#Simulate multiple walks

all_walks=[]

for i in range(500):
    #initialize our random walk
    random_walk=[0]
    
    np.random.seed(124)


    for x in range(100):
       steps = random_walk[-1]
       #simulate a dice roll
       dice_roll = np.random.randint(1,7)
       if dice_roll<=2:
          steps=max(0, steps-1) #to ensure you don't go below zero
       elif dice_roll<=5:
          steps=steps+1
       else:
          steps=steps+(np.random.randint(1,7))
       #implement clumsiness
       if np.random.rand()<=0.001:
            steps = 0
       #add the next step to our random walk
       random_walk.append(steps)

    all_walks.append(random_walk)    
 
    
print(all_walks) #list of lists
```

```python
#convert all_walks to an array
walks_array=np.array(all_walks)
walks_array_t=np.transpose(walks_array)
print(walks_array_t)
```

```python
#visualize all walks
plt.plot(walks_array_t)
plt.title('500 Random Walks')
plt.show()
```

```python
#Plot the distribution of the walks
ends=walks_array_t[-1]
plt.hist(ends)
plt.title('Distribution of 500 walks')
plt.show()
```

```python
#Calculate the odds
count = len(ends[ends>=50])
odds = (count/500) * 100
print(odds)
```

### Conclusion: There odds of reaching 50 steps up the building are high!
